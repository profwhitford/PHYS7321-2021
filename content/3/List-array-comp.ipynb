## arrays vs. lists in Python

here is the default in Python (a list)

```python
IamAlist=[1,2]
```

you can duplicate the list with the command

```python
IamAlist=IamAlist*2
```

```python
IamAlist
```

the list of mutable

```python
IamAlist[2]="A"
```

Here is an array using the **array** module.  The array must be declared using array.array  

```python
import array as arr
```

```python
BasicArray=arr.array("i", [1, 4, 5])
```

a limitation is that you need to specify the type ("i" for integer), and all elements must be the same type

Again, is mutable in the same way....

```python
BasicArray[2]=7
```

Now a NumPy array

```python
import numpy as np
```

elements can be of different types (not allowed by array.array). Array is generated by a list.

```python
NParray = np.array([1.1,2,"A"])
```

np detects and assigns the type

```python
NParray
```

If the elements are the same type, then you can multiply every element by a common factor...

```python
NParray2 = np.array([1.1,2,3.14159])
```

```python
NParray=NParray2*2
```

```python
NParray
```
